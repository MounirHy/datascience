```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()

#this gives us the sum of missing values in each column
```

```python
df.duplicated().sum()
#there are no missing values , when we run the code we get an output of np.int64(0), which means there are 0 duplicated values
```

```python
#we know from before that there are missing values in director , we are going to display the first few rows
df[df["director"].isnull()].head()

#we see under director column NAN
```

```python
df["director"]= df["director"].fillna("unknown")
# in this code we replace the NAN with unknown
```

```python
df["director"].isnull().sum()
#here we verify that are no missing values
```

```python
df[["title","director"]].head()

#here i am displaying the rows to show that the values are replaced by "unknown" , and these steps will be repeated for the rest of the data where needed
```

```python
df["cast"]=df["cast"].fillna("unknown")
df["country"]=df["country"].fillna("unknown")
df["rating"]=df["rating"].fillna("unknown")

# in this cell we replaced the missing values that are big with unknown
```

```python
df=df.dropna(subset=["date_added"])
# we drop the rows with missing dates, they are not much so they wont affect us
```

```python
df=df.dropna(subset=["duration"])
# we drop the rows with missing duration for the same reason
```

```python
# new we make sure we have no more missing values
df.isnull().sum()

# and we can see everthing is 0 , so our work is done with the cleaning
```

```python
df.describe(include="all")
```

```python
df["main_genre"]=df["listed_in"].str.split(",").str[0]

#this code allows us to take the first genre listed and make a new column with the main genre

df[["listed_in","main_genre"]].head()

#and this code is to visualize the first 5 rows
```

```python
df["genre_count"]=df["listed_in"].str.count(",")+1

#this code is to create a calculated columns from the listed genre and to know how many genre this is in each

df[["listed_in","genre_count"]].head(10)
#this code to visualize the first 10 rows , and each displaying the genre and count
```

```python
df.head()
# to make sure the 2 calculated columns are done and see them
```

```python
df["type"].value_counts()
#this shows the type count
```

```python
content_type=df["type"].value_counts()

content_type
#this is to store the data in the variable content type for easier calculation

```

```python
content_type.plot(kind="bar")
plt.title("nb of movies and tv shows")
plt.xlabel("content type")
plt.ylabel("count")

plt.show()

#we can see from the plt.show() the graph and the distribution of movies vs tv shows
```

```python
content_type.plot(kind="pie",autopct="%1.1f%%",ylabel="")

plt.title("movies vs tv shows")
plt.show()
# this is just another way to visualise the distribution
```

```python
country_count = df["country"].value_counts().head(10)

country_count
```

```python
country_count.plot(kind="barh")
plt.title("top 10 countries by nb")
plt.xlabel("nb of titles")
plt.ylabel("country")

plt.show()
# we can see the country with most titles is US
```

```python
genre_count = df["main_genre"].value_counts().head(10)

genre_count
```

```python
genre_count.plot(kind="barh")
plt.title("top 10 genres")
plt.xlabel("nb of titles")
plt.ylabel("country")

plt.show()
```

```python
rating_type = pd.crosstab(df["rating"], df["type"])

rating_type
# this shows a table  comparing the 2 categories

```

```python
rating_type.plot(kind="bar")
plt.title("ratings distribution")
plt.xlabel("rating")
plt.ylabel("nb ")

plt.show()
```

```python
director_count=df[df["director"] != "unknown"]
director_count=director_count["director"].value_counts().head(10)
director_count
```

```python
director_count.plot(kind="barh")

plt.title("top 10 directors")
plt.xlabel("directors")
plt.ylabel("nb")

plt.show()
```

This project analyzed the Netflix Titles dataset using Python and Pandas. The data was cleaned, missing values were handled, and two calculated columns were created to improve the analysis. The results showed that Movies dominate the Netflix catalog, the United States contributes the highest number of titles, Drama is the most common genre, TV-MA is the most frequent rating, and a few directors have a significantly larger number of titles than others. Overall, the project demonstrates how exploratory data analysis can be used to extract meaningful insights from real-world dataset
